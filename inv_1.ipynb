```python
import pandas as pd
```

```python
use_cols = ['NPI', 'Entity Type Code', 'Provider Organization Name (Legal Business Name)',
            'Provider First Name','Healthcare Provider Taxonomy Code_1',
            'Healthcare Provider Taxonomy Code_2',
            'Provider Business Practice Location Address Postal Code'
            'Provider Business Practice Location Address State Name']

codes = [
    '1223G0001X', '1223D0001X', '1223E0200X', '1223P0221X',
    '1223P0300X', '1223P0700X', '1223S0112X', '1223X0400X',
    '122300000X', '1223D0008X', '1223X0008X'
]
```

```python
chunks = []

for chunk in pd.read_csv('../npidata_pfile_20050523-20260308.csv', low_memory=True, usecols=use_cols, chunksize=50000):
    mask = chunk[[c for c in chunk.columns if "taxonomy code_" in c.lower()]].isin(codes).any(axis=1)
    filtered_chunk = chunk[mask]
    if len(filtered_chunk) > 0:
        chunks.append(filtered_chunk)
    print(f"Processed chunk with {len(chunk)} rows, found {len(filtered_chunk)} matches.")

df = pd.concat(chunks, ignore_index=True)
```
