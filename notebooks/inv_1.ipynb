```python
import pandas as pd
```

```python
use_cols = ['NPI', 'Entity Type Code', 'Provider Organization Name (Legal Business Name)',
            'Provider First Name', 'Provider Last Name (Legal Name)', 'Provider First Line Business Mailing Address',
            'Provider First Line Business Practice Location Address',
            'Healthcare Provider Taxonomy Code_1',
            'Healthcare Provider Taxonomy Code_2',
            'Provider Business Practice Location Address Postal Code',
            'Provider Business Practice Location Address State Name']

codes = [
    '1223G0001X', '1223D0001X', '1223E0200X', '1223P0221X',
    '1223P0300X', '1223P0700X', '1223S0112X', '1223X0400X',
    '122300000X', '1223D0008X', '1223X0008X'
]
```

```python
chunks = []

for chunk in pd.read_csv('../npidata_pfile_20050523-20260308.csv', low_memory=False, usecols=use_cols, chunksize=50000):
    mask = chunk[[c for c in chunk.columns if "taxonomy code_" in c.lower()]].isin(codes).any(axis=1)
    filtered_chunk = chunk[mask]
    if len(filtered_chunk) > 0:
        chunks.append(filtered_chunk)
    print(f"Processed chunk with {len(chunk)} rows, found {len(filtered_chunk)} matches.")

df = pd.concat(chunks, ignore_index=True)
```

```python
def fetch_cpb_dental(api_key: str) -> pd.DataFrame:
     url = "https://api.census.gov/data/2017/cbp"
     params = {
          'get': 'ESTAB,EMP,PAYANN,NAME',
          'for': 'county:*',
          'NAICS2017': ['62121','621210'],
          'key': api_key
     }
     response = requests.get(url, params=params)
     response.raise_for_status()

     data = response.json()

     df = pd.DataFrame(data[1:], columns=data[0])
     return df
```

```python
def fetch_gcspi() -> pd.DataFrame:
    #requests pull: https://www.newyorkfed.org/medialibrary/research/interactives/GSCPI/downloads/gscpi_data.xlsx 
    #pressure index only value column, positive value indicates above average supply chain pressure
    url = "https://www.newyorkfed.org/medialibrary/research/interactives/GSCPI/downloads/gscpi_data.xlsx"
    resp = requests.get(url)
    df = pd.read_excel(resp.content)
    return df[['Date', 'GSCPI']]
```
